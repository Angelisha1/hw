# Задача №1

```python
import pandas as pd
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
ratings = pd.read_csv('ml-latest-small/ratings.csv')

joined_tables = ratings.merge(movies, on = 'movieId', how = 'left')

highest_rating = joined_tables.groupby('movieId').count().sort_values('rating', ascending = False).filter(items = ['movieId'])

title = highest_rating.merge(movies, on = 'movieId', how = 'left')
title.head()
```

# Задача №2

```python
file = pd.read_csv('power.csv')

baltic = file[ (file['country'] == 'Latvia') | (file['country'] == 'Estonia') | (file['country'] == 'Lithuania')]

year = baltic [ baltic['year'].between(2005, 2010) ]

categories = year[ (year['category'] == 4) | (year['category'] == 12) | (year['category'] == 21)  ]

quantity_filter = categories [categories.quantity >= 0]

total_sum = quantity_filter.quantity.sum()

print(total_sum)
```
